```python
import folium
import pandas as pd
import branca.colormap as cm
from pyproj import Geod

# =========================
# GEODESIC ENGINE
# =========================
geod = Geod(ellps="WGS84")

def create_map(gdf, positive_tags, negative_tags):
    df = gdf.copy()
    
    # Initialize map
    m = folium.Map(location=[df.y.mean(), df.x.mean()], zoom_start=18, tiles=None)

    # Google Hybrid with 0.5 opacity
    folium.TileLayer(
        tiles="https://mt1.google.com/vt/lyrs=y&x={x}&y={y}&z={z}",
        attr="Google",
        name="Google Hybrid",
        overlay=True,
        opacity=0.5
    ).add_to(m)

    # Colormaps
    pos_cmap = cm.LinearColormap(["red", "yellow", "green"], vmin=0, vmax=10)
    neg_cmap = cm.LinearColormap(["green", "yellow", "red"], vmin=0, vmax=10)

    # Create FeatureGroups
    groups = {t: folium.FeatureGroup(name=t, show=(t == 'score')).add_to(m) 
              for t in ['score'] + positive_tags + negative_tags}

    # Helper to draw arrow with black border
    def draw_arrow(group, lat, lon, heading, color, tt, html_popup):
        lon2, lat2, _ = geod.fwd(lon, lat, heading, 10)
        adj_h = heading + 30
        
        # 1. Draw Black Border Arrow
        folium.PolyLine([(lat, lon), (lat2, lon2)], color='black', weight=7, opacity=1.0).add_to(group)
        folium.RegularPolygonMarker((lat2, lon2), number_of_sides=3, radius=8, rotation=adj_h, 
                                    color='black', fill_color='black', fill_opacity=1).add_to(group)
        
        # 2. Draw Colored Arrow
        folium.PolyLine([(lat, lon), (lat2, lon2)], color=color, weight=5, opacity=1.0).add_to(group)
        folium.RegularPolygonMarker((lat2, lon2), number_of_sides=3, radius=6, rotation=adj_h, 
                                    color=color, fill_color=color, fill_opacity=1).add_to(group)

    # Drawing function
    def add_to_layer(group, row, lat, lon, heading, val, cmap, tt, html_popup):
        color = cmap(val)
        
        # Draw Arrow FIRST
        if pd.notna(heading):
            draw_arrow(group, lat, lon, heading, color, tt, html_popup)
            
        # Draw Point LAST (So it renders on top of the arrow)
        folium.CircleMarker(
            (lat, lon), radius=8, color='black', 
            fill=True, fill_color=color, fill_opacity=1.0, 
            opacity=1.0, weight=1, tooltip=tt,
            popup=folium.Popup(html_popup, max_width=500)
        ).add_to(group)

    # Main Loop
    for _, row in df.iterrows():
        lat, lon = row.y, row.x
        h = row.heading
        
        # Tooltip text
        tt = f"<b>Score: {row.get('score', 0):.2f}</b><br><u><b>Pos:</b></u><br>"
        tt += "<br>".join([f"{t}: {row.get(t,0):.2f}" for t in positive_tags])
        tt += "<br><u><b>Neg:</b></u><br>"
        tt += "<br>".join([f"{t}: {row.get(t,0):.2f}" for t in negative_tags])
        
        # Popup HTML
        img_path = row.get('path', '')
        html_popup = f"""
        <div style="display: flex; flex-direction: row; align-items: flex-start; width: 400px;">
            <div style="flex: 1; margin-right: 10px; font-size: 12px;">{tt}</div>
            <div style="flex: 1;">
                <a href="{img_path}" target="_blank">
                    <img src="{img_path}" style="width:150px; border:1px solid #ccc; cursor:pointer;">
                </a>
            </div>
        </div>
        """
        
        # Add to layers
        add_to_layer(groups['score'], row, lat, lon, h, row.get('score', 0), pos_cmap, tt, html_popup)
        
        for t in positive_tags:
            if pd.notna(row.get(t)):
                add_to_layer(groups[t], row, lat, lon, h, row[t], pos_cmap, tt, html_popup)
        for t in negative_tags:
            if pd.notna(row.get(t)):
                add_to_layer(groups[t], row, lat, lon, h, row[t], neg_cmap, tt, html_popup)

    folium.LayerControl(collapsed=False).add_to(m)
    return m
```
